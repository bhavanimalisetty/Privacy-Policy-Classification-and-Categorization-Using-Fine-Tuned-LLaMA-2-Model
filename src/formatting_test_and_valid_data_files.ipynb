```python

import pandas as pd
import json
import os
# Load the JSON file
with open('../Datasets/opp115/train.json', 'r') as file:
    data_valid_raw = json.load(file)

# Prepare a list to hold the expanded data
expanded_data = []

# Instruction text that will be common for all entries
instruction_text = "Given a privacy policy text, classify it into one of these categories: First Party Collection/Use, Third Party Sharing/Collection, Other, User Choice/Control, Do Not Track, International and Specific Audiences, Data Security, Policy Change, Data Retention, User Access, Edit and Deletion"

# Expand the data
for item in data_valid_raw:
    text = item['text']
    labels = item['label']
    for label in labels:
        expanded_data.append({
            'instruction': instruction_text,
            'input': text,
            'output': label
        })

# Convert the list to a DataFrame
df = pd.DataFrame(expanded_data)

# Define the path for saving the file in the parent directory
output_path = "../Datasets/Formatted_Data_Sets/train_formatted.csv"
df.to_csv(output_path, index=False)

```

```python
import pandas as pd

# Sample DataFrame creation or loading here
df = pd.read_csv("../Datasets/Formatted_Data_Sets/train_formatted.csv")

text_col = []
for _, row in df.iterrows():
    # Constructing the prompt that combines the instruction, input, and output in the required format
    prompt = "Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request."
    instruction = str(row["instruction"])
    input_query = str(row["input"])
    response = str(row["output"])
    if len(input_query.strip())==0:
        text = prompt + "### Instruction:\n" + instruction +"\n ### Response: \n " + response
    else:
        text = (prompt + "### Instruction:\n" + instruction + "\n ###Input:\n "+ input_query + "\n ### Response: \n " + response)
    text_col.append(text)

df.loc[:,"text"] = text_col
output_path = "../Datasets/Formatted_Data_Sets/train.csv"
df.to_csv(output_path, index=False)
```
